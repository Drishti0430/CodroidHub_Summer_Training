## Graph comparing Year and Price with color distinctions for different Fuel_Type

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from CSV
file_path = 'used_cars_data.csv'  # Provide the correct path to your file
data = pd.read_csv(file_path)

# Drop rows with missing values in Year and Price
data_cleaned = data.dropna(subset=['Year', 'Price', 'Fuel_Type'])

# Plot the data
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_cleaned, x='Year', y='Price', hue='Fuel_Type', palette='viridis', s=100)

# Add titles and labels
plt.title('Used Car Prices vs Year (by Fuel Type)', fontsize=14)
plt.xlabel('Year of Manufacture', fontsize=12)
plt.ylabel('Price (Lakhs)', fontsize=12)
plt.legend(title='Fuel Type')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
for col in num_var:
    print(col)
    print('Skew :',round(data[col].skew(),2))
    plt.figure(figsize=(8,3))
    plt.subplot(1,2,1)
    data[col].hist(grid=False)
    plt.ylabel('Count')
    plt.subplot(1,2,2)
    sns.boxplot(x=data[col])
    plt.show()
```

# What are the insights of these graphs?

### from graph-1 : Price Trends Over Time:

- You can observe how prices vary with the year of manufacture. Generally, newer cars are expected to be more expensive.
  Impact of Fuel Type:
- By using different colors for fuel types, you can see if certain fuel types tend to be priced differently or if there are noticeable trends within each fuel type.

### From graph-2 : Comparative Analysis:

- Comparing histograms and boxplots of different variables can reveal differences in distribution, central tendency, and presence of outliers across variables.
